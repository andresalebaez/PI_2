- Importar las librerías necesarias para el análisis de datos:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from scipy import stats
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

- Cargaremos los datos y los organizaremos en un DataFrame de Pandas:

```python
IBM_df = pd.read_csv('C:\\Users\\andres\\AppData\\Local\\Programs\\Python\\Python310\\PI-2\\IBM.csv')
INTC_df = pd.read_csv('C:\\Users\\andres\\AppData\\Local\\Programs\\Python\\Python310\\PI-2\\INTC.csv')
MSFT_df = pd.read_csv('C:\\Users\\andres\\AppData\\Local\\Programs\\Python\\Python310\\PI-2\\MSFT.csv')
```

```python
data = IBM_df[['Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume']]
```

```python
corr_matrix = data.corr()
```

```python
sns.heatmap(corr_matrix, annot=True, cmap="YlGnBu")
plt.show()
```

```python
# Crear las variables x e y para el análisis de regresión
x = data['Open'].values
y = data['Close'].values

# Realizar el análisis de regresión lineal simple
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

# Crear la línea de regresión
line = slope * x + intercept

# Crear la gráfica de dispersión y la línea de regresión
plt.scatter(x, y)
plt.plot(x, line, color='red')

# Añadir las etiquetas de los ejes y el título
plt.xlabel('Open')
plt.ylabel('Close')
plt.title('Análisis de regresión lineal simple')

# Mostrar la gráfica
plt.show()
```

```python
future_price = slope*IBM_df[['Open', 'High', 'Low', 'Close', 'Volume']] + intercept
print(future_price)
```

```python
plt.plot(future_price)
plt.xlabel('Período de tiempo')
plt.ylabel('Precio de la acción')
plt.title('Predicción del precio de la acción')
plt.show()
```
