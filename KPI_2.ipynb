```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
# Descarga los datos históricos de Yahoo Finance para IBM, INTC y MSFT
ibm = yf.download('IBM')
intc = yf.download('INTC')
msft = yf.download('MSFT')

# Calcula la media móvil de 50 días para el precio de cierre de cada empresa
ibm['SMA50'] = ibm['Close'].rolling(window=50).mean()
intc['SMA50'] = intc['Close'].rolling(window=50).mean()
msft['SMA50'] = msft['Close'].rolling(window=50).mean()

# Imprime las últimas filas del dataframe de cada empresa con el KPI SMA50
print('IBM')
print(ibm.tail())
print('INTC')
print(intc.tail())
print('MSFT')
print(msft.tail())
```

```python
# Grafica el promedio móvil de 50 días para cada empresa
plt.plot(ibm['Close'], label='IBM')
plt.plot(ibm['SMA50'], label='IBM SMA50')
plt.plot(intc['Close'], label='INTC')
plt.plot(intc['SMA50'], label='INTC SMA50')
plt.plot(msft['Close'], label='MSFT')
plt.plot(msft['SMA50'], label='MSFT SMA50')

# Agrega leyendas y título al gráfico
plt.legend()
plt.title('Promedio móvil de 50 días para IBM, INTC y MSFT')
plt.xlabel('Fecha')
plt.ylabel('Precio de cierre')

# Muestra el gráfico
plt.show()

```

```python
# Calcula el retorno diario para cada empresa
ibm['Daily Return'] = ibm['Adj Close'].pct_change(1)
intc['Daily Return'] = intc['Adj Close'].pct_change(1)
msft['Daily Return'] = msft['Adj Close'].pct_change(1)

# Imprime las últimas filas del dataframe de cada empresa con el KPI Daily Return
print('IBM')
print(ibm.tail())
print('INTC')
print(intc.tail())
print('MSFT')
print(msft.tail())
```

```python
# Grafica el retorno diario para cada empresa
plt.plot(ibm['Daily Return'], label='IBM')
plt.plot(intc['Daily Return'], label='INTC')
plt.plot(msft['Daily Return'], label='MSFT')

# Agrega leyendas y título al gráfico
plt.legend()
plt.title('Retorno diario para IBM, INTC y MSFT')
plt.xlabel('Fecha')
plt.ylabel('Retorno diario')

# Muestra el gráfico
plt.show()
```

```python
# Calcula el volumen promedio para cada empresa durante los últimos 30 días
ibm['Average Volume'] = ibm['Volume'].rolling(window=30).mean()
intc['Average Volume'] = intc['Volume'].rolling(window=30).mean()
msft['Average Volume'] = msft['Volume'].rolling(window=30).mean()

# Imprime las últimas filas del dataframe de cada empresa con el KPI Average Volume
print('IBM')
print(ibm.tail())
print('INTC')
print(intc.tail())
print('MSFT')
print(msft.tail())
```

```python
# Grafica el volumen promedio para cada empresa
plt.plot(ibm['Average Volume'], label='IBM')
plt.plot(intc['Average Volume'], label='INTC')
plt.plot(msft['Average Volume'], label='MSFT')

# Agrega leyendas y título al gráfico
plt.legend()
plt.title('Volumen promedio para IBM, INTC y MSFT')
plt.xlabel('Fecha')
plt.ylabel('Volumen promedio')

# Muestra el gráfico
plt.show()
```
